```python
def plot_cert(path,name,title):
    legends = []
    for g in glob(path):
        print(g)
        legends.append(g[-13:-4])
        f = open(str(g), "r")
        radius, correct = [], []
        count = 0

        for line in f:
            if count > 0:
                _,_,_,c,d,_ = line.split('\t')
                radius.append(float(c))
                correct.append(int(d))
            count += 1

        rad = np.array(radius)*np.array(correct)
        axs = np.linspace(0, max(radius) + 0.02, 1000)
        cer = []
        for r in axs:
            cer.append((rad > r).sum())
        cer = np.array(cer)/len(rad)
        plt.plot(axs,cer)
        plt.xlabel('Radius')
        plt.ylabel('Certified Acc.')
        plt.grid()
        plt.legend(legends)
        plt.title(title)
        plt.savefig('./plots/certified_acc'+name+'.png')
    plt.close()
```

```python
plot_cert("experiments/experiments/pred_smoothing/clustr_only/clustr_sigma_*",'Augmentation', 'smoothing sigma = 0.1')
```

```python
for g in glob('experiments/experiments/pred_smoothing/clustr_only/*/samples4'):
    print(g[-16:])
    title = g[-16:-9] + '_' + g[-8:]
    name = g[-16:-9] + '_' + g[-8:]
    plot_cert(g+'/*', name, title)
```
